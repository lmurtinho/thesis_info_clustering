# Experiment on information-theoretic clustering (dataset Newsgroups 20)

This notebook allows for the reproduction (running, plotting, and storing) of the experiments presented in "Theoretical and experimental results on information-theoretical clustering", Lucas Murtinho's Master's thesis. The experiments concern the Partition with Minimum Weighted Entropy Problem (PMWEP).

## Import modules

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
import tests
```

```python
plt.rcParams['figure.figsize'] = [18,10]
```

## Select data set

Three data sets are available for testing:

- Newsgroups 20
- RCV1
- A synthetic data set 

For more information, please refer to section 6.1 of the thesis. For different data sets to be analyzed in this notebook, they should be stored as csv files in the same folder as the notebook.

```python
data_set = "ng20"
# data_set = "poisson"
```

## Run full tests

The cell below runs tests for the methods selected. Additional parameters include:

- `k_list`: a list with the numbers of clusters to be tested
- `seed_list`: a list with the seed being used for testing
- `m_list`: a list with the sizes of the coreset used in the coreset model
- `max_iter`:  a list with the maximum number of iterations for iteration-based models (coreset, Lloyd's algorithm, and divisive clustering)
- `verbose`: whether to present information on the tests as they are run

```python
res = tests.test_results(data_names=[data_set],
                         headers=[False],
                         libname="libclust.so",
                         methods=["rg2_clustering",
                                  "lc_clustering",
                                  "di_clustering",
                                  "rd_clustering",
                                  "bh_clustering",
                                  "star_clustering"],
                         k_list=[2, 5, 10, 20, 30, 40, 50, 100, 200, 300, 400, 
                                500, 750, 1000, 1250, 1500, 1750, 2000],
                         seed_list=[1, 2, 3, 4, 5],
                         m_list=[5000],
                         max_iter_list=[0, 100],
                         verbose=False)
df = pd.DataFrame.from_dict(res)
```

## Function to get method from data point

The function below retrieves the method being used for each data point in the data frame produced in the cell above.

```python
def get_method(s):
    method = s['method']
    if not np.isnan(s['max_iter']):
        method += '_' + str(int(s['max_iter']))
    if not np.isnan(s['m']):
        method += '_' + str(int(s['m']))
    return method

df['full_method'] = df.apply(get_method, axis=1)
```

## Function to change method's names

Helper function to change the names of the methods, from the ones in the C library to the ones used in the dissertation.

```python
def change_names(s):
    fm = s['full_method']
    fm = fm.replace('lc_clustering', 'coreset')
    fm = fm.replace('star_clustering', 'star')
    fm = fm.replace('rg2_clustering', 'ratio_greedy')
    fm = fm.replace('di_clustering', 'divisive')
    fm = fm.replace('bh_clustering', 'lloyd')
    fm = fm.replace('rd_clustering', 'random')
    return fm

df['full_method'] = df.apply(change_names, axis=1)
```

## Pivot tables

Creates pivot tables for the entropy and running time of each method in the data frame.

```python
pivot_entropy = pd.pivot_table(df, values='entropy', columns='full_method', index='k')
pivot_time = pd.pivot_table(df, values='time', columns='full_method', index='k')
```

## Plot average entropy per model (full version of iteration-based models)

```python
columns = ['ratio_greedy', 'star', 'lloyd_100', 
           'coreset_100_5000', 'divisive_100']
ax = sns.scatterplot(data=pivot_entropy[columns], s = 200)
ax.set_title("20 Newsgroups - Average entropy per model", fontsize=24);
ax.legend(fontsize=20, markerscale=2);
ax.set_xlabel('k', fontsize=20);
ax.set_ylabel('Average entropy', fontsize=20);
ax.tick_params(axis='both', which='major', labelsize=20)

plt_name = data_set + "_entropy.png"
plt.savefig(plt_name)
```

## Plot average running time per model (full version of iteration-based models)

```python
columns = ['ratio_greedy', 'star', 'lloyd_100', 
           'coreset_100_5000', 'divisive_100']
ax = sns.scatterplot(data=pivot_time[columns], s = 200)
ax.set_title("20 Newsgroups - Average running time per model", fontsize=24);
ax.legend(fontsize=20, markerscale=2);
ax.set_xlabel('k', fontsize=20);
ax.set_ylabel('Average running time (seconds)', fontsize=20);
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set(yscale='log');

plt_name = data_set + "_time.png"
plt.savefig(plt_name)
```

## Plot average entropy per model (initial version of iteration-based models)

```python
columns = ['ratio_greedy', 'star', 'lloyd_0', 
           'coreset_0_5000', 'divisive_0']
ax = sns.scatterplot(data=pivot_entropy[columns], s = 200)
ax.set_title("20 Newsgroups - Average entropy per model", fontsize=24);
ax.legend(fontsize=20, markerscale=2);
ax.set_xlabel('k', fontsize=20);
ax.set_ylabel('Average entropy', fontsize=20);
ax.tick_params(axis='both', which='major', labelsize=20)

plt_name = data_set + "_entropy_initial.png"
plt.savefig(plt_name)
```

## Plot average running time per model (initial version of iteration-based models)

```python
columns = ['ratio_greedy', 'star', 'lloyd_0', 
           'coreset_0_5000', 'divisive_0']
ax = sns.scatterplot(data=pivot_time[columns], s = 200)
ax.set_title("20 Newsgroups - Average running time per model", fontsize=24);
ax.legend(fontsize=20, markerscale=2);
ax.set_xlabel('k', fontsize=20);
ax.set_ylabel('Average running time (seconds)', fontsize=20);
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set(yscale='log');

plt_name = data_set + "_time_initial.png"
plt.savefig(plt_name)
```

## Save results to csv

```python
csv_full = data_set + "_full.csv"
csv_entropy = data_set + "_entropy.csv"
csv_time = data_set + "_time.csv"

pivot_entropy.to_csv(csv_entropy, float_format="%.6f")
pivot_time.to_csv(csv_time, float_format="%.6f")
df.to_csv(csv_full, float_format="%.6f")
```
